#### <b>What are Python Loops?</b>
If you need to do something 100 times, you certainly don't need to write it out in 100 identical lines of code.

In Python, loops allow you to iterate over a sequence, whether that's a list, tuple, string, or dictionary.

A loop allows you to repeat actions without rewriting the same code.

#### <b>What types of Python Loops are there?</b>
There is a <b>for</b> loop and a <b>while</b> loop.
For Loop: Used when you know how many times you want to iterate.
While Loop: Used when the number of iterations depends on a condition.

#### <b>The range() function</b>
In most cases, we will want to loop through a predetermined list. However, we can also generate a list to loop over "on the fly" with the <b>range()</b> function.

````python
for i in range(8):
    print(i)
````

You can use the <b>range()</b> function together with the <b>len()</b> function to loop through a list, along with a variable that has the current index in the sequence.

````python
my_list = [10, 20, 30, 40, 50, 60]

for i in range(len(my_list)):
    print(i, my_list[i])
````

#### <b>Enumerate</b>
You can achieve the same effect as above by using the enumerate() function, to get two values each step through the loop.


````python
# the enumerate function will return two values each time the loop statement is run:
# "indx" will be the current index value, and "val" will be the current element of my_list
for indx, val in enumerate(my_list):
    print(indx, val)
````

# Loop Control Statement ( Break & Continue)

1. Using break in a Loop
The break statement is used to exit a loop prematurely when a certain condition is met.

````python
# Finding the First Even Number in a List

numbers = [1, 3, 5, 6, 7, 9]

for num in numbers:
    if num % 2 == 0:
        print(f"First even number found: {num}")
        break
print("End of loop.")
````

Explanation: The loop stops as soon as it finds the first even number.

2. Using continue in a Loop.
- The continue statement skips the rest of the code inside the loop for the current iteration and moves to the next iteration.

````python
# Skipping Odd Numbers
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num % 2 != 0:
        continue  # Skip odd numbers
    print(f"Even number: {num}")
````
